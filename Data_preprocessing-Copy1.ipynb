```python
# drop columns, previously index from data2, duplicates - track_name_y, size_bytes_y
data = data.drop('Unnamed: 0', axis=1)
data.shape
```

```python
def drop_duplicate_columns(data):
    """Drops columns with duplicate data - track_name_y, size_bytes_y, 
    only if all values are same for duplicate columns"""
    
    track_name_y, size_bytes_y = True, True
    
    for index, row in data.iterrows():
        if row['track_name_x'] != row['track_name_y']:
            track_name_y = False
    
    for index, row in data.iterrows():
        if row['size_bytes_x'] != row['size_bytes_y']:
            size_bytes_y = False

    if track_name_y:
        data = data.drop('track_name_y', axis=1)
    if size_bytes_y:
        data = data.drop('size_bytes_y', axis=1)
        
    return data

data = drop_duplicate_columns(data)
data.shape
```
